```python
import pandas as pd
import numpy as np
import json
import os
```

```python
df_cat1_dev = pd.read_excel('./result/textcnn_pred_res-v5.xlsx')
df_cls_dev = pd.read_csv('./result/structbert_pred_res-v3.csv')
```

```python
# config
TRAIN_DIR = './data/wenshu_ms_dataset/train/'
TEST_DIR = './data/wenshu_ms_dataset/test/'
DEV_DIR = './data/wenshu_ms_dataset/dev/'
RESULT_DIR = './result/'
DATA_DIR = './data/wenshu_ms_dataset/'


# TRAIN_FILES = sorted(os.listdir(TRAIN_DIR), key=lambda x:int(x[:-5]))
# TEST_FILES = sorted(os.listdir(TEST_DIR), key=lambda x:int(x[:-5]))
DEV_FILES = sorted(os.listdir(DEV_DIR), key=lambda x:int(x[:-5]))

RECLASS_TYPES = ['公司事务', '劳动人事', '合同事务', '婚姻家庭', '房地产纠纷', '民事纠纷'] 
```

```python
from sklearn.preprocessing import MinMaxScaler
import copy
def Normalized(df_):
    df = copy.deepcopy(df_)
    Scaler = MinMaxScaler(feature_range=(0, 1))
    df.iloc[:, 1:] = Scaler.fit_transform(df.iloc[:, 1:].values.T).T
    return df
```

```python
# 归一化
Scaler = MinMaxScaler(feature_range=(0, 1))
df_cat1_dev.iloc[:, 2:] = Scaler.fit_transform(df_cat1_dev.iloc[:, 2:].values.T).T
df_cat1_dev.set_index('q_id', inplace=True, drop=False)
df_cat1_dev
```

```python
df_cat_dev = copy.deepcopy(df_cat1_dev)
df_cat_dev.set_index('q_id', inplace=True, drop=False)
df_cat_dev
```

```python
ctxs_cat = []
# dev_data
for k, file in enumerate(DEV_FILES):
    with open(DEV_DIR + file, 'r', encoding='utf-8') as f:
        js_dt = json.load(f)
        tmp_cat = []
        for idx, case in js_dt['ctxs'].items():
            cat1 = case['Category']['cat_1']
            cat2 = case['Category']['cat_2']
            tmp_cat.append([cat1+'-'+cat2])
        ctxs_cat.append(tmp_cat)
        
np.array(ctxs_cat).shape
```

```python
df_cls_group = df_cls_dev.groupby('q_id')
```

```python
ctxs_score_cal = []
for i in range(len(ctxs_cat)):
    q_id = df_cat_dev['q_id'].tolist()[i]
    tmp_score_cal = []
    df_tmp = df_cls_group.get_group(q_id).set_index('c_id')
    for idx, ctxs in enumerate(ctxs_cat[i]):
        score1 = df_cat1_dev.loc[q_id, ctxs[0]]
        score3 = df_tmp.loc[idx, 'label']
#         score4 = df_tmp.loc[idx, '1'] if score3 == 1 else df_tmp.loc[idx, '0'] 
        tmp_score_cal.append([score1, score3])
    ctxs_score_cal.append(tmp_score_cal)
np.array(ctxs_score_cal).shape
```

```python
# rule 1
ans_lst = []
for ctxs in ctxs_score_cal:
    tmp_df = pd.DataFrame(ctxs, columns = ['category', 'cls'])
    ans_lst.append(tmp_df.sort_values(by=['category', 'cls'], ascending=False).index.tolist())
```

```python
q_ids = [str(x) for x in df_cat_dev['q_id'].values]
res_dict = dict(zip(q_ids, ans_lst))
res_dict
```

```python
# 打包
class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        if isinstance(obj, np.floating):
            return float(obj)
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        return super(NpEncoder, self).default(obj)

result_path = RESULT_DIR + 'dev_ans_v5-5.json'
with open(result_path, 'w') as fp:
    json.dump(res_dict, fp=fp, ensure_ascii=False, cls=NpEncoder, indent=4)
    fp.close()
```
